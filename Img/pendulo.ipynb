```python
import numpy as np   
import matplotlib.pyplot as plt 
```

```python
theta_0 = np.pi/4.0 
g = 9.8
l = 1.0
w = np.sqrt(g/l)
t = np.linspace(0,10,20)
theta = theta_0*np.cos(w*t)
x = l*np.sin(theta)
y = -l*np.cos(theta)
plt.figure(figsize=(10,10))
#plt.plot(x,y)
plt.scatter(0,0)
plt.scatter(x,y)
plt.show()
```
